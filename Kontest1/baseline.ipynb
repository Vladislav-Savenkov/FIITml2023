```python
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
```

```python
df_train = pd.read_csv('train.csv')
X = df_train.drop('Target', axis=1)
y = df_train['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.21)

df_train.dropna(axis=0, how='all', inplace=True)
df_train.dropna(axis=1, how='all', inplace=True)
df_train.head()
```

```python

df_train.head()
```

```python
clf = CatBoostClassifier(eval_metric='AUC', use_best_model=True)

cat_features = X.select_dtypes(include=['object']).columns.tolist()
#cat_features = df_train.columns[df_train.dtypes == 'object'].tolist()
clf.fit(X_train, y_train, cat_features=cat_features, eval_set=(X_test, y_test))
```

```python
df_to_predict = pd.read_csv('data_predict.csv')

y_pred = clf.predict_proba(df_to_predict)[:, 1]
print(y_pred)
```

```python
submission = pd.DataFrame({'ID': df_to_predict['ID'].to_list(), 'Target': y_pred})
```
